# **Download Video and Extract Audio:**
The function download_video takes a URL as input, downloads the corresponding video using the pytube library, and saves it to a specified output path. Then, the extract_audio function extracts the audio from the downloaded video using moviepy library and saves it as a WAV file.

````python
!pip install pytube
````

````python
from pytube import YouTube
from moviepy.editor import VideoFileClip
import os

def download_video(url, output_path='/content/video.mp4'):
    yt = YouTube(url)
    stream = yt.streams.filter(file_extension='mp4').first()
    stream.download(filename=output_path)
    return output_path

def extract_audio(video_path, audio_path='/content/audio.wav'):
    video_clip = VideoFileClip(video_path)
    video_clip.audio.write_audiofile(audio_path)
    video_clip.close()
    return audio_path
````

# Transcription of Audio:
  For transcription, you're using an open-source Speech-to-Text model from Hugging Face's Transformers library, specifically the "openai/whisper-base" model. This model is pre-trained on LibriSpeech dataset and fine-tuned on additional data by OpenAI. It transcribes the audio into text in chunks of specified duration, returning the transcription along with timestamps.


- **Explanation of Chosen Model and Techniques:**
The ```openai/whisper-base``` model is selected due to its good performance in speech recognition tasks. It's capable of handling various accents and noisy environments, making it suitable for general-purpose use. Additionally, it supports GPU acceleration for faster processing, enhancing its efficiency.

- we could use better models like ```openai/whisper-large``` or ```openai/whisper-medium``` etc. but they are very bulky to implement over colab and take a long time to load and werent really required as the base model does the job just fine

- Also this model can give the transcription with the time-stamps which takes care Time-Alignment of Transcript with audio by itself

````python
!pip install transformers pipeline torch datasets
````

````python
import torch
from transformers import pipeline
from datasets import load_dataset
````

````python
def transcribe(audio_path, chunk_duration=15):
  device = "cuda:0" if torch.cuda.is_available() else "cpu"
  pipe = pipeline(
    "automatic-speech-recognition",
    model="openai/whisper-base",
    chunk_length_s= chunk_duration,
    device=device,
  )
  ds = load_dataset("hf-internal-testing/librispeech_asr_dummy", "clean", split="validation")
  results = pipe(audio_path, batch_size= 128, return_timestamps=True)
  return results
````

Here, ```batch_size= 128``` means that the model will process 128 samples at a time. This can help speed up the transcription process by making better use of the resources available. By processing multiple inputs simultaneously, larger batch sizes can take advantage of parallelism, leading to more efficient use of computational resources like GPUs. This can reduce the overall time required for inference.

larger batch sizes can improve throughput (total amount of work done in a given time)

Larger batch sizes require more memory. If the batch size is too large, it can exceed the available memory on your GPU or CPU (runtime crashed on colab when i used batch size higher than 128). Beyond a certain point, increasing the batch size might not lead to much improvements in speed due to factors like hardware limitations.

The ```chunk_length_s``` parameter specifies the length of audio chunks that the model processes at a time, but it **does not guarantee that the transcription output will align perfectly with these chunk boundaries.** The model might generate text and timestamps based on its internal understanding and processing of the audio, which can result in variable segment lengths. semantic chunking is done in the comming steps

# Semantic Chunking of Data:
The ```create_semantic_chunks``` function slices the transcription into smaller, manageable chunks of text, with each chunk having a duration of less than 15 seconds.

  - **If a segment of text exceeds the maximum chunk length, it is split into smaller subchunks to ensure each chunk meets the duration requirement.**

  - Otherwise, **consecutive segments are added into a single chunk until the maximum duration is reached.**

**Logic Used for Semantic Chunking:**
- Segment Length Check: Each segment's length is checked against the maximum chunk length. If it exceeds the limit, it's split into smaller subchunks to ensure the chunk duration constraint.

- Adding: If a segment can be added to the current chunk without exceeding the maximum duration, it's appended to the current chunk. if not, a new chunk is started.

**Strengths:**

- Utilizes both text semantics and audio characteristics for chunking, which enhances the coherence of chunks.
- Handles long segments (that are longer than 15secs) by splitting them into smaller, manageable parts.

**Weaknesses:**

- Relies heavily on the accuracy of the transcription model. Errors in transcription could cause errors in chunking.

- The splitting logic based on word boundaries may not always preserve the natural flow of speech (normally people have pauses and changing rate of words).

````python
def create_semantic_chunks(transcription_result, max_chunk_length=15.0):
    chunks = []
    current_chunk = {
        "chunk_id": 1,
        "chunk_length": 0.0,
        "text": "",
        "start_time": 0.0,
        "end_time": 0.0,
    }

    for i, segment in enumerate(transcription_result):
        segment_text = segment['text']
        segment_start = segment['timestamp'][0]
        segment_end = segment['timestamp'][1] if segment['timestamp'][1] is not None else segment_start + max_chunk_length

        segment_length = segment_end - segment_start

        if segment_length > max_chunk_length:
            # Split the long segment into smaller chunks

            # number of chunks that ca be made will be length of the big chunk/15.0, convert it to integer and add 1
            num_subchunks = int(segment_length // max_chunk_length) + 1

            #length of each subchunk
            subchunk_length = segment_length / num_subchunks

            #we iterate throught the chunks
            for j in range(num_subchunks):

                #start of j'th chunk
                subchunk_start = segment_start + j * subchunk_length

                #end of j'th chunk
                subchunk_end = subchunk_start + subchunk_length

                # (the main part) text for the j'th subchunk will be the defined by indexing the segment text from the begining of jth part of the segment text and the begining of the next subchunk
                #this relies heavily on the audio. assuming that the audio was countinuous and had contant rate throughout the large chunk
                subchunk_text = segment_text.split()[int(j * len(segment_text.split()) / num_subchunks) : int((j + 1) * len(segment_text.split()) / num_subchunks)]

                #add the chunk to the chunks dictionary
                subchunk_text = ' '.join(subchunk_text)
                chunks.append({
                    "chunk_id": len(chunks) + 1,
                    "chunk_length": subchunk_length,
                    "text": subchunk_text,
                    "start_time": subchunk_start,
                    "end_time": subchunk_end,
                })
        else:
            # Check if adding this segment would exceed the maximum chunk length(i.e 15.0)
            if current_chunk['chunk_length'] + segment_length <= max_chunk_length:
                # we add the segment to the current chunk
                if current_chunk['text']:
                    current_chunk['text'] += " " + segment_text
                else:
                    current_chunk['text'] = segment_text
                current_chunk['end_time'] = segment_end
                current_chunk['chunk_length'] += segment_length
            else:
                # we finish the current chunk and start a new chunk
                chunks.append(current_chunk)
                current_chunk = {
                    "chunk_id": len(chunks) + 1,
                    "chunk_length": segment_length,
                    "text": segment_text,
                    "start_time": segment_start,
                    "end_time": segment_end,
                }

    # Add the last chunk if it has content
    if current_chunk['text']:
        chunks.append(current_chunk)

    return chunks
````

# **Gradio-app Interface**

````python
!pip install gradio
````

````python
def process(url):
    # download the video
    video_path = download_video(url) #downloads the video and returns path  '/content/video.mp4'

    #extract the audio
    audio_path = extract_audio(video_path) #extract audio from video the video and returns path  '/content/audio.wav'

    #get the transtriptions with time stamps
    transcriptions = transcribe(audio_path) # returns chunks with irregular chunk lengths

    #semantic chunking
    chunks = create_semantic_chunks(transcriptions['chunks']) #returns the transcriptions in the required format

    #return the chunks
    return chunks

# function that creats a single string output to be shown in the gradio output box
def format_output(transcriptions):
    formatted_output = ""
    for chunk in transcriptions:
        formatted_output += f"Chunk id : {chunk['chunk_id']}\n chunk length : {chunk['chunk_length']}\n Text: {chunk['text']} \n start_time :{chunk['start_time']}s \n end_time : {chunk['end_time']}s \n\n"
    print(formatted_output)
    return formatted_output.strip() #removes any leading or trailing whitespaces

def transcribe_interface(url):
    #getting the transcrubtions in the chunks dictionary format
    chunks = process(url)

    #returning the transcruption for the gradio output box
    return format_output(chunks)
````

````python
import gradio as gr
myapp = gr.Interface(
    fn=transcribe_interface,
    inputs=[gr.Textbox(placeholder = "video URL", show_label = False)],
    outputs= gr.Textbox(label="Transcription"),
    title="YouTube Video Transcriber",
)
myapp.launch(debug=True)
````

````python
import os
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import requests

chrome_options = Options()
chrome_options.add_argument("--headless")# so that it doesnt show on screen
chrome_options.add_argument("--no-sandbox")#disables Chrome's security mechanisms
chrome_options.add_argument("--disable-dev-shm-usage")#Disables using the /dev/shm shared memory file system. reduces memory usage during testing
driver = webdriver.Chrome(options=chrome_options)#creating session

#Making the list webpages containing book name and number of chapters
webpages = [['MAT',28], ['MRK',16], ['LUK', 24], ['JHN', 21], ['ACT', 28], ['ROM', 16], ['1CO', 16],
            ['2CO', 13], ['GAL', 6], ['EPH', 6], ['PHP',4], ['COL',4], ['1TH',5], ['2TH', 3], ['1TI', 6],
            ['2TI', 4], ['TIT', 3], ['PHM', 1], ['HEB',13], ['JAS', 5], ['1PE', 5], ['2PE', 3], ['1JN', 5],
            ['2JN', 1], ['3JN', 1], ['JUD', 1], ['REV', 22]]

#making lists for data that couldn't be scrapped due to connection timeout
audio_miss = []
text_miss = []

#---------------------------Function to scrape Text-----------------------------
def text_scrape(book, chapter):
  url = f"https://live.bible.is/bible/HINOHC/{book}/{chapter}?audio_type=audio"
  txt_name = f'/content/text/{book}{chapter}.txt'


  driver.get(url)

  try:
    #WebDriverWait is used to wait fr the element to get loaded in the webpage. it takes in an integer(20secs here) as inpu too so that it does not wait infinitely for that element to load
    span_elements = WebDriverWait(driver, 20).until(
            EC.presence_of_all_elements_located((By.CLASS_NAME, 'align-left'))
    )

    #concatinating all the hindi text into a single string (they were in parts on the webpage in different span tags)
    text = " ".join([element.text for element in span_elements])

    #creating a text file in writing mode and saving the sting in that txt file and then saving the text file
    with open(txt_name, 'w') as file:
      file.write(text.strip())
  except Exception as e: #indicating that scrapping of that text file failed
    print(f"text scrapping failed for({book}{chapter})")
    text_miss.append([book, chapter]) #stored in text_miss list for scrapping again later

#-------------------------------Function to scrape Audio------------------------
def audio_scrape(book, chapter):
  url = f"https://live.bible.is/bible/HINOHC/{book}/{chapter}?audio_type=audio"
  aud_name = f'/content/audio/{book}{chapter}.mp3'

  driver.get(url)

  try:
    #WebDriverWait is used to wait fr the element to get loaded in the webpage. it takes in an integer (40secs here) as inpu too so that it does not wait infinitely for that element to load
    audio_element = WebDriverWait(driver, 40).until(
            EC.presence_of_element_located((By.CLASS_NAME, 'audio-player'))
    )

    #getting the source of the audio form the src attribute of the video tag with class .audio-player
    audio_src = audio_element.get_attribute('src')

    #downloading/fetching the audio
    audio = requests.get(audio_src, stream=True)#downloading audio

    #creating a binary file and writing the audio in it and saving it .mp3 file
    with open(aud_name, 'wb') as f:
      for chunk in audio.iter_content(chunk_size=None):
        f.write(chunk)
  except Exception as e: #indicating that scrapping of that audio file failed
    print(f"Audio scrapping failed for({book}{chapter})")
    audio_miss.append([book, chapter]) #stored in audio_miss list for scrapping again later

for i in webpages:
  for j in range(1, i[1]+1, 1):
    if(os.path.isfile(f'/content/text/{i[0]}{j}.txt')):
      print(f'text:({i[0]}, {j}) already exists')
    else:
      text_scrape(i[0], j)

    if(os.path.isfile(f'/content/audio/{i[0]}{j}.mp3')):
      print(f'audio:({i[0]}, {j}) already exists')
    else:
      audio_scrape(i[0], j)

driver.quit()#ending the session
````

**INSERTING AUDIO FEATURES INTO THE DATAFRAME**
- **RMS (Root Mean Square)**
  - RMS is a measure of the amplitude of the audio signal. It provides an estimate of the loudness or energy of the signal.
  - It is commonly used to measure the average power of the audio signal over time.

- **ZCR (Zero Crossing Rate)**
  - ZCR is the rate at which the audio signal changes sign (crosses the zero amplitude line) within a given time frame.
  - It is used to analyze the noisiness or the frequency content of the signal. Higher ZCR values indicate more high-frequency content or noise.
  - computed by counting the number of times the signal changes sign.

- **Spectral Centroid**
  - The spectral centroid is a measure that indicates where the center of mass of the spectrum is located. It is often perceived as the brightness of a sound.
  - It is used to describe the timbre of the sound. Higher spectral centroids are often associated with clear and crisp speech sounds.  important for distinguishing between phonemes, especially consonants

- **Spectral Bandwidth**
  - Spectral bandwidth measures the width of the spectrum. It quantifies the spread of the frequencies around the spectral centroid.
  - It is used to describe the timbre and texture of the sound. Wide spectral bandwidth indicates rich and complex sounds, often found in vowel sounds.helps in better phoneme recognition
  - calculated as the square root of the variance of the spectrum.

````python
import librosa
import numpy as np
def calculate_audio_attributes(audio_path):
    y, sr = librosa.load(audio_path)
    duration = librosa.get_duration(y=y, sr=sr)
    rms = np.mean(librosa.feature.rms(y=y))
    zcr = np.mean(librosa.feature.zero_crossing_rate(y=y))
    spectral_centroid = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr))
    spectral_bandwidth = np.mean(librosa.feature.spectral_bandwidth(y=y, sr=sr))

    return duration, rms, zcr, spectral_centroid, spectral_bandwidth

df[['duration', 'rms', 'zcr', 'spectral_centroid', 'spectral_bandwidth']] = df['audio_path'].apply(lambda x: pd.Series(calculate_audio_attributes(x)))
df
````

**SNR, HNR, and Spectral Flatness**
- **Signal-to-Noise Ratio (SNR)**
  - High SNR: Clearer audio with less background noise.
    - STT: More accurate transcriptions.
    - TTS: Clearer, more intelligible speech synthesis.
  - Low SNR: More background noise.
    - STT: Reduced transcription accuracy.
    - TTS: Lower quality synthesized speech.
- **Harmonics-to-Noise Ratio (HNR)**
  - High HNR: High harmonic content, less noise.
    - STT: Better phoneme recognition.
    - TTS: Natural and pleasant synthesized speech.
  - Low HNR: More noise relative to harmonics.
    - STT: Poorer phoneme recognition.
    - TTS: Unnatural, noisy synthesized speech.
- **Spectral Flatness**
  - Low Spectral Flatness: Tone-like signal with clear harmonics.
    - STT: Clear speech aids accurate recognition.
    - TTS: Natural-sounding speech.
  - High Spectral Flatness: Noise-like signal.
    - STT: Harder to distinguish phonemes.
    - TTS: Robotic, less natural speech.

````python
import pandas as pd
import numpy as np
import librosa
import seaborn as sns
import matplotlib.pyplot as plt

def calculate_snr(audio, sr):
    signal_power = np.mean(audio**2)
    noise_power = np.mean((audio - np.mean(audio))**2)
    snr = 10 * np.log10(signal_power / noise_power)
    return snr

# Flatness
def calculate_spectral_flatness(audio, sr):
    spectral_flatness = librosa.feature.spectral_flatness(y=audio)
    return np.mean(spectral_flatness)

# HNR
def calculate_hnr(audio, sr):
    harmonic, percussive = librosa.effects.hpss(audio)
    hnr = np.mean(librosa.amplitude_to_db(np.abs(harmonic), ref=np.max) -
                  librosa.amplitude_to_db(np.abs(percussive), ref=np.max))
    return hnr

# Processing audio files
def process_audio(file_path):
    audio, sr = librosa.load(file_path)
    # Calculating all metrics
    snr = calculate_snr(audio, sr)
    spectral_flatness = calculate_spectral_flatness(audio, sr)
    hnr = calculate_hnr(audio, sr)

    return snr, spectral_flatness, hnr

# Add audio clarity metrics to the dataframe
clarity_metrics = {'snr': [], 'spectral_flatness': [], 'hnr': []}
for index, row in df.iterrows():
    print(index)
    audio_path = row['audio_path']
    snr, spectral_flatness, hnr = process_audio(audio_path)
    df.at[index, 'snr'] = snr
    df.at[index, 'spectral_flatness'] = spectral_flatness
    df.at[index, 'hnr'] = hnr


# Visualize the results
plt.figure(figsize=(15, 10))
plt.subplot(3, 1, 1)
sns.histplot(df['snr'], kde=True)
plt.title('SNR Distribution')

plt.subplot(3, 1, 2)
sns.histplot(df['spectral_flatness'], kde=True)
plt.title('Spectral Flatness Distribution')

plt.subplot(3, 1, 3)
sns.histplot(df['hnr'], kde=True)
plt.title('HNR Distribution')

plt.tight_layout()
plt.show()
````
